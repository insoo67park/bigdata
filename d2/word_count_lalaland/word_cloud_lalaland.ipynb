```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline

#  
f = open("/home/insoo67_park/word_count_lalaland/word_count_output_lalaland", 'r')
lines = f.readlines()

word_count_result = dict() # word_cout    
for line in lines: # line    
    
    # UTF-8 ,  , ":"   
    line = line.strip().split(":") 
    
    word =''.join(line[:-1]) # count     
    count = int(line[-1]) # count 
    word_count_result[word] = count # {keyword: count} 
f.close()

# word cloud 
wordcloud = WordCloud(
    font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf',
    width = 800,
    height = 800,
    background_color="white",
)

# word cloud 
wordcloud = wordcloud.generate_from_frequencies(word_count_result)
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
